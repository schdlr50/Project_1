3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:

* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# функция для перевода длительности опыта работы в месяца
column_expa=hh_database['Опыт работы']
exp_years_words=["года", "лет", "год"] # список слов для поиска целого года
exp_months_words=["месяца", "месяц", "месяцев"] # список слов для поиска месяцев
def get_exp(arg):
    if arg is np.nan or arg=='Не указано':
        return None
    arg_split=arg.split(' ')[:8]
    months=0 #выставляем на значение на 0
    years=0 #выставляем на значение на 0
    for index, item in enumerate(arg_split):
        if item in exp_years_words: 
            years=int(arg_split[index-1])
        if item in exp_months_words:
            months=int(arg_split[index-1])
    return int(years*12+months)
hh_database['Опыт работы(месяц)']=hh_database['Опыт работы'].apply(get_exp)
# смотрим медианное значение
display(hh_database['Опыт работы(месяц)'].median())
# убираем столбец 'Опыт работы'
hh_database=hh_database.drop(columns=['Опыт работы'], axis=1)
```

```python
#функция по выделению города из признака 'Город, переезд, командировки'
def get_city(arg):
    column_city=arg.split(' , ')[0]
    cities_million=['Новосибирск','Екатеринбург','Нижний Новгород','Казань','Челябинск','Омск','Самара','Ростов-на-Дону','Уфа','Красноярск','Пермь','Воронеж','Волгоград']
    if column_city in cities_million:
        return 'Город-миллионник'
    elif column_city == "Москва":
        return 'Москва'
    elif column_city == "Санкт-Петербург":
        return 'Санкт-Петербург'
    else:
        return 'Другие'
hh_database['Город']=hh_database['Город, переезд, командировки'].apply(get_city)
#процент соискателей по городу:
count_spb=len(hh_database[hh_database['Город']=='Санкт-Петербург']) #кол-во соискателей, проживающих в питере
count_total=len(hh_database['Город']) #общее кол-во соискателей
print(round(count_spb/count_total*100))
    
#функция для формирования признака "Готовность к переезду"
def get_move_to(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg):
        return False
    else:
        return True
hh_database['Готовность к переезду']=hh_database['Город, переезд, командировки'].apply(get_move_to)

#функция для формирования признака "Готовность к командировкам"
def get_buisness_trip(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else:
            return True
    else:
        return False
#процент готовых к переезду и командировкам
hh_database['Готовность к командировкам']=hh_database['Город, переезд, командировки'].apply(get_buisness_trip)
print(round(hh_database[hh_database['Готовность к командировкам'] & hh_database['Готовность к переезду']].shape[0]/hh_database.shape[0]*100))
#удаляем признак 'Город, переезд, командировки'
hh_database=hh_database.drop(columns=['Город, переезд, командировки'], axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# создаем списки возможных вариантов занятости и графиков
employment=('частичная занятость', 'проектная работа', 'полная занятость', 'стажировка', 'волонтерство')
chart=('гибкий график', 'полный день', 'сменный график', 'вахтовый метод', 'удаленная работа')
# функция для присваивания True или False в зависимости от варианта
for employment, chart in zip(employment, chart):
    hh_database[employment] = hh_database['Занятость'].apply(lambda x: employment in x)
    hh_database[chart] = hh_database['График'].apply(lambda x: chart in x)
# кол-во людей, которые ищут проектную работу и волонтёрство
print(((hh_database['волонтерство']==True) & (hh_database['проектная работа']==True)).value_counts())
# кол-во людей, которые хотят работать вахтовым методом и с гибким графиком
print(((hh_database['вахтовый метод']==True) & (hh_database['гибкий график']==True)).value_counts())
# удаляем признаки 'Занятость' и 'График'
hh_database=hh_database.drop(columns=['Занятость', 'График'], axis=1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# переводим признак 'обновление резюме' в формат datetime, убираем время
hh_database['Обновление резюме']=pd.to_datetime(hh_database['Обновление резюме']).dt.date
# переводим признак 'date' в формат datetime
exchangeRates=pd.read_csv('data/ExchangeRates.csv')
exchangeRates['date']=pd.to_datetime(exchangeRates['date']).dt.date
# уберем лишнюю информацию из таблицы exchangeRates
exchangeRates=exchangeRates.drop(columns=['per', 'time', 'vol'], axis=1)
# выделяем сумму ЗП и наименование валюты в новом датасэте
def find_zp(arg): 
    zp = float(arg.split(' ')[0])
    return zp
def find_curr(arg):
    curr=arg.split(' ')[1]
    return curr
# копируем основной датасэт для преобразований 
df_ZP=hh_database.copy()
# производим преобразование с признаками
df_ZP['сумма зп']=hh_database['ЗП'].apply(find_zp)   
df_ZP['валюта зп']=hh_database['ЗП'].apply(find_curr)
# убираем признак 'ЗП' из основного датасэта
hh_database=hh_database.drop(columns=['ЗП'], axis=1)
# создаем словарь для дальнейшего приведения стандарта
currency ={'руб.':'RUB','KZT':'KZT','USD':'USD','бел.руб.':'BYN','EUR':'EUR','грн.':'UAH','сум':'UZS','KGS':'KGS','AZN':'AZN'} 
# приводим во вспомогательном датасэте к стандарту значение валют
df_ZP['валюта зп']=df_ZP['валюта зп'].replace(currency)
# приcоединяем к таблице с резюме таблицу с курсами валют
merge_df=df_ZP.merge(exchangeRates,left_on=['валюта зп', 'Обновление резюме'],right_on=['currency', 'date'],how='left')
# зададим для валюты RUS значение в признаке close
merge_df['close'].fillna(1, inplace=True)
merge_df['proportion'].fillna(1, inplace=True)
# умножаем сумму желаемой заработной платы на курс валюты (close) и делим на пропорцию
merge_df['ЗП(руб.)']=round(merge_df['сумма зп']*merge_df['close']/merge_df['proportion'])
merge_df['ЗП(руб.)'].median()
# переносим признак 'ЗП(руб.)' в основной датасет
hh_database['ЗП(руб.)']=merge_df['ЗП(руб.)']
# сверямся 
hh_database.info()
```
